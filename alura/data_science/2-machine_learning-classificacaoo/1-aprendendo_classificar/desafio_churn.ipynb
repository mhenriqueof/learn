## One Hot Encoder

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = make_column_transformer((
    OneHotEncoder(drop= 'if_binary'), ['sexo_biologico', 'pais', 'tem_cartao_credito', 'sexo_biologico']),
    remainder= 'passthrough', sparse_threshold= 0
)
```

```python
x_columns = x.columns
x_columns
```

```python
x_ohe = ohe.fit_transform(x)
```

```python
ohe.get_feature_names_out(x_columns)
```

```python
pd.DataFrame(x_ohe, columns= ohe.get_feature_names_out(x_columns))
```

## Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
y_le = le.fit_transform(y)
```

```python
y_le
```

# 3

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_ohe, y_le, stratify= y_le, random_state= 5)
```

```python
from sklearn.dummy import DummyClassifier
dummy = DummyClassifier()
dummy.fit(x_train, y_train)
dummy.score(x_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(max_depth= 5, random_state= 5)
tree.fit(x_train, y_train)
y_score = tree.score(x_test, y_test)
y_score
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(tree, filled= True, class_names= ['no', 'yes']);
```

# 4

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
norm = MinMaxScaler()
x_test_norm = norm.fit_transform(x_test)
x_train_norm = norm.fit_transform(x_train)
```

```python
pd.DataFrame(x_train_norm)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
knn.fit(x_train_norm, y_train)
knn.score(x_test_norm, y_test)
```

```python
print(f'Acurácia Dummy: {dummy.score(x_test, y_test)}')
print(f'Acurácia Árvore: {tree.score(x_test, y_test)}')
print(f'Acurácia KNN: {knn.score(x_test_norm, y_test)}') 
```
