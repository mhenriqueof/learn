```python
def check_outliers(coluna, df):
    
    Q1 = df[f'{coluna}'].quantile(.25)
    Q3 = df[f'{coluna}'].quantile(.75)
    IQR = Q3 - Q1
    lower_fence = Q1 - 1.5 * IQR
    upper_fence = Q3 + 1.5 * IQR
    
    outliers_index = (df[f'{coluna}'] < lower_fence) | (df[f'{coluna}'] > upper_fence)
    
    return print(df[outliers_index])
```

```python
check_outliers('Valor da transação', data)
```
