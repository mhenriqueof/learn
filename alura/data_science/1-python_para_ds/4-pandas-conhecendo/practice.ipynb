```python
import pandas as pd

df = pd.DataFrame({
   'Animal': ['Cachorro', 'Gato', 'Elefante', 'Cachorro', 'Gato', 'Elefante'],
   'Cor': ['Preto', 'Branco', 'Cinza', 'Marrom', 'Preto', 'Marrom'],
   'Quantidade': [2, 3, 1, 4, 2, 2]
})
df
```

```python
dataset = df.groupby('Cor')[['Quantidade']].sum()
dataset
```

```python
dataset.plot(kind= 'bar',
             figsize= (15,10),
             color = 'black')
```
