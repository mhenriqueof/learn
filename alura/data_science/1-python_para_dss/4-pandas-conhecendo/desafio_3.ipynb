```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
```

```python
data = pd.read_csv(url)
data
```

```python
data.isnull().sum()
```

```python
df1_na = data.fillna(0)
df1_na
```

```python
alunos = df1_na.query('"Alice" in Nome | Nome == "Carlos"').index
df2_alunos = df1_na.drop(alunos, axis= 0)
df2_alunos
```

```python
df2_alunos = df2_alunos.reset_index()
df2_alunos
```

```python
df2_alunos = df2_alunos.drop(columns= ['index'], axis= 1)
df2_alunos
```

```python
aprovados = df2_alunos['Aprovado'] == True
df3_aprovados = df2_alunos[aprovados]
df3_aprovados
```

```python
df4_notas = df3_aprovados.copy()
df4_notas['Notas'].replace(to_replace= 7.0, value= 8.0, inplace= True)
df4_notas
```

```python
df3_aprovados
```

```python
df3_aprovados.to_csv('alunos_aprovados.csv')
```
