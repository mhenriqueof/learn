# Manipulando os dados

## Criando colunas numéricas

```python
import pandas as pd
data = pd.read_csv(filepath_or_buffer= 'aluguel.csv',
                   sep= ';')
data
```

```python
data.rename(columns= {'Valor': 'Aluguel'}, 
            inplace= True)
data
```

```python
data['Valor/mes'] = (data['Aluguel'] + data['Condominio']).round()
data
```

```python
data.rename(columns= {'Valor/mes': 'Valor/Mes'},
            inplace = True)
data
```

```python
data['Valor/Ano'] = ((data['Valor/Mes'] * 12) + data['IPTU']).round()
data
```

## Criando colunas categóricas

```python
data['Descricao'] = data['Tipo'] + ' em ' + data['Bairro'] + '. Possui ' + data['Quartos'].astype(str) + ' quarto(s) e ' + data['Vagas'].astype(str) + ' vaga(s) de garagem.'
data
```

```python
data['Possui_Suites'] = data['Suites'].apply(lambda valor: "Sim" if valor > 0 else "Não")
data.head()
```
