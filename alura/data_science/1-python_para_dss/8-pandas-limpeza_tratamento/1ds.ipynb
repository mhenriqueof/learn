Cliente

genero: gênero (masculino e feminino) \
idoso: informação sobre um(a) cliente ter ou não idade igual ou maior que 65 anos \
parceiro: se o(a) cliente possui ou não um(a) parceiro ou parceira\
dependentes: se o(a) cliente possui ou não dependentes\
tempo_servico: meses de contrato do(a) cliente\
\
Serviço de telefonia

servico_telefone: assinatura de serviço telefônico\
varias_linhas: assinatura de mais de uma linha de telefone\

Serviço de internet\
\
servico_internet: assinatura de um provedor internet\
seguranca_online: assinatura adicional de segurança online\
backup_online: assinatura adicional de backup online\
protecao_dispositivo: assinatura adicional de proteção no dispositivo\
suporte_tecnico: assinatura adicional de suporte técnico, menos tempo de espera\
tv_streaming: assinatura de TV a cabo\
filmes_streaming: assinatura de streaming de filmes\
\
Conta

contrato: tipo de contrato\
faturamente_eletronico: se o(a) cliente prefere receber a fatura online\
metodo_pagamento: forma de pagamento\
cobranca.mensal: total de todos os serviços do(a) cliente por mês\
cobranca.Total: total gasto pelo(a) cliente

## Normalizar

```python
import json
```

```python
with open("data/dataset-telecon.json") as file:
    gross_json = json.load(file)
```

```python
norm_data = pd.json_normalize(gross_json)
norm_data.sample(5)
```

```python
norm_data.info()
```

## Tratar strings vazias

```python
norm_data[norm_data['conta.cobranca.Total'] == ' ']
```

```python
norm_data[norm_data['conta.cobranca.Total'] == ' '][['cliente.tempo_servico', 'conta.contrato', 'conta.cobranca.mensal', 'conta.cobranca.Total']]
```

```python
idx = norm_data[norm_data['conta.cobranca.Total'] == ' '].index
idx
```

```python
norm_data.loc[idx, 'conta.cobranca.Total'] = norm_data.loc[idx, 'conta.cobranca.mensal'] * 24
```

```python
norm_data.loc[idx, 'cliente.tempo_servico'] = 24.0
```

```python
norm_data.loc[idx, :]
```

```python
norm_data = norm_data.astype({'conta.cobranca.Total': 'float64'})
```

```python
norm_data.info()
```

```python
type_data = norm_data.copy()
```

## Outliers

```python
outliers_data_6 = null_data_5.copy()
```

```python
outliers_data_6.describe()
```

```python
import seaborn as sns
```

```python
sns.boxplot(x= outliers_data_6['cliente.tempo_servico'])
```

```python
import plotly.express as px
```

```python
fig = px.box(data_frame= outliers_data_6, x= 'cliente.tempo_servico')
fig.show()
```

```python
Q1 = outliers_data_6['cliente.tempo_servico'].quantile(.25)
Q3 = outliers_data_6['cliente.tempo_servico'].quantile(.75)
IQR = Q3 - Q1
lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR
```

```python
print(Q1)
print(Q3)
print(IQR)
print(lower_fence)
print(upper_fence)
```

```python
outliers_index = (outliers_data_6['cliente.tempo_servico'] < lower_fence) | (outliers_data_6['cliente.tempo_servico'] > upper_fence)
outliers_index
```

```python
outliers_data_6[outliers_index][['cliente.tempo_servico']]
```

```python
outliers_data_6.loc[outliers_index, 'cliente.tempo_servico'] = ceil(
    outliers_data_6.loc[outliers_index, 'conta.cobranca.Total'] / outliers_data_6.loc[outliers_index, 'conta.cobranca.mensal']
)
```

```python
fig = px.box(outliers_data_6, x= 'cliente.tempo_servico')
fig.show()
```

```python
outliers_data_6[outliers_index][['cliente.tempo_servico']]
```

```python
Q1 = outliers_data_6['cliente.tempo_servico'].quantile(.25)
Q3 = outliers_data_6['cliente.tempo_servico'].quantile(.75)
IQR = Q3 - Q1
lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR
```

```python
outliers_index_2 = (outliers_data_6['cliente.tempo_servico'] < lower_fence) | (outliers_data_6['cliente.tempo_servico'] > upper_fence)
outliers_index_2
```

```python
outliers_data_6[outliers_index_2]
```

```python
outliers_data_6 = outliers_data_6[~outliers_index_2]
```

```python
outliers_data_6.reset_index(drop= True, inplace= True)
outliers_data_6
```

```python
fig = px.box(outliers_data_6, x= 'cliente.tempo_servico')
fig.show()
```
