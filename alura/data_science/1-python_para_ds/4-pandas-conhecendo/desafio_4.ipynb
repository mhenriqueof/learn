```python
import pandas as pd
```

```python
data = pd.read_csv(filepath_or_buffer= 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv')
data
```

```python
df = data.copy().fillna(0)
df
```

```python
df['Situação'] = df['Aprovado'].apply(lambda logico: "Aprovado" if logico == True else "Reprovado")
df
```

```python
df.drop(columns= ['Aprovado'], axis= 1, inplace= True)
df
```

```python
df['Pontos_Extra'] = df['Notas'].apply(lambda nota: nota * 0.4)
df = df.iloc[:, [0, 1, 2, 4, 3]]
df
```

```python
df['Notas_Finais'] = df['Notas'] + df['Pontos_Extra']
df
```

```python
df['Aprovados_Final'] = df['Notas_Finais'] >= 6
df
```

```python
reprovados1 = df.query('Situação == "Reprovado"')
reprovados1
```

```python
aprovados = reprovados1.query('Aprovados_Final == True')
aprovados
```
