```python
import sqlalchemy
```

```python
from sqlalchemy import create_engine, MetaData, table, inspect
```

```python
engine = create_engine('sqlite:///:memory:')
```

```python
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv')
data
```

```python
data.to_sql('clients', engine, index= False)
```

```python
inspector = inspect(engine)
print(inspector.get_table_names())
```

```python
query = "SELECT * \
         FROM clients \
         WHERE Categoria_de_renda = 'Empregado'"
```

```python
employees = pd.read_sql(query, engine)
employees
```

```python
employees.to_sql('employees', con= engine, index= False)
```

```python
pd.read_sql_table('employees', con= engine, columns= ['Idade', 'Estado_civil'])
```

```python
query_2 = 'SELECT *\
           FROM clients'
```

```python
query_3 = 'DELETE\
            FROM clients\
                WHERE ID_Cliente=5008804'
with engine.connect() as conn:
    conn.execute(query_3)
```
