```python
import pandas as pd
```

```python
data = pd.read_csv('data/dados_inadimplencia.csv')
data
```

---

```python
data.describe()
```

```python
data['inadimplente'].value_counts()
```

---

```python
x = data.drop(columns='inadimplente')
x
```

```python
y = data['inadimplente']
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, stratify=y)
```

---
**Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
dtc_model = DecisionTreeClassifier(max_depth=5, random_state=42)
dtc_model.fit(x_train, y_train)
y_pred_dtc = dtc_model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_dtc))
```

---
**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
pipeline = make_pipeline(StandardScaler(), LogisticRegression())
pipeline.fit(x_train, y_train)
y_pred_lr = pipeline.predict(x_test)
```

```python
print(classification_report(y_test, y_pred_lr))
```

---
**Grid Search CV**

```python
from sklearn.model_selection import GridSearchCV
import numpy as np
```

```python
param_grid_dtc = {
    'criterion': ['gini', 'entropy'],
    'max_depth': np.linspace(5, 15, 5, dtype=int),
    'min_samples_split': np.linspace(5, 20, 4, dtype=int),
    'min_samples_leaf': np.linspace(5, 20, 4, dtype=int),
    'max_features': ['sqrt', 'log2', 'log10'],
    'splitter': ['best', 'random']
}
```

```python
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold(shuffle=True, random_state=42)
```

```python
dtc_grid_search = GridSearchCV(estimator=DecisionTreeClassifier(random_state=42),
                               param_grid=param_grid_dtc,
                               scoring="recall", 
                               cv=cv)
```

```python
dtc_grid_search.fit(x_train, y_train)
```

```python
dtc_grid_search.best_params_
```

```python
df_dtc_grid_results = pd.DataFrame(dtc_grid_search.cv_results_)
df_dtc_grid_results.head()
```

```python
df_dtc_grid_results.sort_values(by='mean_test_score', ascending=False).head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(
    df_dtc_grid_results['param_max_depth'],
    df_dtc_grid_results['mean_test_score'],
    )

plt.title('max_depth vs. mean_test_score')
plt.xlabel('max_depth')
plt.ylabel('recall')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(
    df_dtc_grid_results['param_min_samples_leaf'],
    df_dtc_grid_results['mean_test_score'],
    )

plt.title('param_min_samples_leaf vs. mean_test_score')
plt.xlabel('param_min_samples_leaf')
plt.ylabel('recall')
plt.show()
```
