# 3. Agrupamentos multi-index

```python
df_per_sector = df_year.groupby(['Gás', 'Nível 1 - Setor'])[['Emission']].sum()
df_per_sector
```

```python
df_per_sector.xs(('CO2 (t)', 'Mudança de Uso da Terra e Floresta'), level= [0, 1])
```

```python
table_sum = df_per_sector.groupby(level= 0).idxmax()
table_sum
```

```python
emission = df_per_sector.groupby(level= 0).max().values
emission
```

```python
table_sum.insert(1, 'Emission values', emission)
table_sum
```

```python
swap_df_per_sector = df_per_sector.swaplevel(0, 1)
swap_df_per_sector
```

```python
swap_df_per_sector.groupby(level= 0).idxmax()
```
