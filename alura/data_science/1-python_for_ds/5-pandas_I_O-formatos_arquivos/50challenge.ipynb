```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv')
data
```

```python
engine = create_engine('sqlite:///:memory:')
```

```python
data.to_sql('challenge5', engine, index= False)
```

```python
query = 'UPDATE challenge5\
            SET Rendimento_anual = 300000\
                WHERE ID_Cliente = 6840104'
with engine.connect() as conn:
    conn.execute(query)
```

```python
pd.read_sql_table('challenge5', engine)
```
